```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("/content/superstore_final_dataset (1)(in).csv", encoding="latin-1")
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
df.head()
```

```python
# Display the summary of the dataset
print("\nSummary of the dataset:")
print(df.info())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
#Checking if any duplicate data exists
df.duplicated().sum()
```

```python
null_rows = df[df.isnull().any(axis=1)]

# Display the rows
null_rows
```

```python
# Removing country and year to make order_id in int
# Extract the last part of Order_ID (after the second '-')
df["Order_ID"] = df["Order_ID"].apply(lambda x: int(x.split("-")[-1]))
```

```python
print(df['Order_ID'].head())
```

```python
df["Customer_ID"] = df["Customer_ID"].apply(lambda x: int(x.split("-")[-1]))
```

```python
print(df['Customer_ID'].head())
```

```python
df["Product_ID"] = df["Product_ID"].apply(lambda x: int(x.split("-")[-1]))
```

```python
print(df['Product_ID'].head())
```

```python
# Drop the Postal_Code column
df.drop(columns=["Postal_Code"], inplace=True)

# Verify the column has been removed
df.columns
```

```python
# Convert Dates to Datetime Format
df["Order_Date"] = pd.to_datetime(df["Order_Date"], dayfirst=True)
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"], dayfirst=True)
```

```python
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
# Drop the 'Country' column
df_cleaned = df.drop(columns=["Country"],inplace=True)

# Verify the column is removed
df.shape, "Country" not in df.columns
```

```python
#Univariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns

# Univariate Analysis - Checking distribution of numerical variables
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df["Sales"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Sales Distribution")

# Checking missing values
missing_values = df.isnull().sum()

# Display the plot
plt.tight_layout()
plt.show()

# Show missing values
missing_values[missing_values > 0]
```

```python
# Drop unnecessary columns
columns_to_drop = ["Row_ID", "Order_ID", "Customer_ID", "Product_ID", "Customer_Name"]
# Drop unnecessary columns in the original dataset
df.drop(columns=columns_to_drop, inplace=True)

# Verify changes
df.info(), df.head()
```

```python
df.head()
```

```python
#Outlier Analysis
#Negative or Unrealistic Sales Values – All sales values are non-negative, meaning there are no clear data entry errors.
# Extreme Outliers (Top 1% Sales Values):
# The highest sales values (e.g., $3,083, $8,159, $3,347) are for high-priced furniture, technology, and office machines.
#These do not appear to be errors, but they may need further review.
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot size and style
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")

# Box plot to visualize outliers in Sales
sns.boxplot(x=df['Sales'])

# Title and labels
plt.title("Box Plot of Sales Values (Outlier Detection)", fontsize=14)
plt.xlabel("Sales", fontsize=12)

# Show the plot
plt.show()

# Box Plot Analysis:
# The box plot confirms significant outliers in Sales.
# The extreme values lie far beyond the whiskers, indicating high-value transactions rather than errors.
```

```python

# Remove the Log_Sales column
df.drop(columns=['Log_Sales'], inplace=True, errors='ignore')

# Save the dataset without Log_Sales column
df.to_csv("superstore_final_cleaned.csv", index=False)

# Provide the download link
from google.colab import files
files.download("superstore_final_cleaned.csv")
```
